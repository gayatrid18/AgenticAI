```python
import os
from dotenv import load_dotenv

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.chat_models import ChatOllama
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
```

```python
loader1=PyPDFLoader("data/AI_Startup.pdf")
documents1 = loader1.load()
loader2=PyPDFLoader("data/HR_Policies_Handbook.pdf")
documents2 = loader2.load()
documents=documents1+documents2
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
docs=text_splitter.split_documents(documents)
print("Total Chunks:",len(docs))
```

```python
embedinngs_openai = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
vectorstore_openai=FAISS.from_documents(docs,embedinngs_openai)
vectorstore_openai.save_local("vectorstore_openai")
```

```python
embeddings_gemini = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python
vectorstore_gemini = FAISS.from_documents(docs,embeddings_gemini)
vectorstore_gemini.save_local("vectorstore_gemini")
```

```python
embeddings_ollama = OllamaEmbeddings(model="embeddinggemma")
```

```python
vectorstore_ollama=FAISS.from_documents(docs, embeddings_ollama)
vectorstore_ollama.save_local("vectorstore_ollama")
```
